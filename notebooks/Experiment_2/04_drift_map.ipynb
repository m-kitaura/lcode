# Visualization of drift

```python
SAVE_TO_FILE = False
SAVE_EXT = '.eps'
```

```python
detector = 'lcode_wd'
DATASETS = DATASETS[:5]
```

## Drift map

```python
for dataset in DATASETS:
    print(dataset)
    agg = AGG(SAVE_DATA_DIR, dataset, detector)
    result = agg.select_results(**{k: v[0] for k, v in DETECTOR_PRMSS[detector].items()})
    logfile_path = result['record'].values[0]
    change_points = DRIFT_POINTS[dataset]
    
    mode = 'summary'
    save_path = os.path.join(OUTPUT_DIR, 'drift_map_%s_%s' % (dataset, mode))
    df, plt = show_drift_map(logfile_path, mode, change_points=change_points)
    fix_plot(plt, True, None, None, SAVE_TO_FILE, save_path, SAVE_EXT)
    
    mode = 'feature'
    save_path = os.path.join(OUTPUT_DIR, 'drift_map_%s_%s' % (dataset, mode))
    df, plt = show_drift_map(logfile_path, mode, change_points=change_points)
    fix_plot(plt, True, None, None, SAVE_TO_FILE, save_path, SAVE_EXT)
    
    mode = 'class'
    save_path = os.path.join(OUTPUT_DIR, 'drift_map_%s_%s' % (dataset, mode))
    df, plt = show_drift_map(logfile_path, mode, change_points=change_points)
    fix_plot(plt, True, None, None, SAVE_TO_FILE, save_path, SAVE_EXT)
    
    mode = 'detected'
    save_path = os.path.join(OUTPUT_DIR, 'drift_map_%s_%s' % (dataset, mode))
    df, plt = show_drift_map(logfile_path, mode, change_points=change_points)
    if plt:
        fix_plot(plt, True, None, None, SAVE_TO_FILE, save_path, SAVE_EXT)
```

# Class ratio of datasets

```python
mode = 'area'
for dataset in DATASETS:
    print(dataset)
    dataset_path = os.path.join(INTERIM_DATA_DIR, dataset+'_cat.csv')
    change_points = DRIFT_POINTS[dataset]
    save_path = os.path.join(OUTPUT_DIR, 'class_ratio_%s_%s' % (dataset, mode))
    df, plt = show_class_ratio(dataset_path, mode, change_points)
    fix_plot(plt, True, None, None, SAVE_TO_FILE, save_path, SAVE_EXT)
    print(df.mean(axis=1))
```
